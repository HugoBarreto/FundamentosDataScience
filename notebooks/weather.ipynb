```python
def get_weather(api_key, city, time, units, lang):
    latitude = city[0]
    longitude = city[0]
    url = "https://api.darksky.net/forecast/{}/{},{},{}?units={}&lang={}".format(api_key, latitude, longitude, time, units, lang)
    r = requests.get(url)
    data = json.loads(r.text)
```

```python
time= "1561950000" #2019-7-1-0-0-0
units = "si"
lang = "en"
daily = 60*60*24
days_in_month = 31
days_in_month = days_in_month-1
```

```python
city=montevideo_geo
for i in range(days_in_month)[0]:
    time_iter = time+i*daily
    day_weather = get_weather(api_key, city, time, units, lang)
    for hour in day_weather:
        hour = data['hourly']['data']
    
```
