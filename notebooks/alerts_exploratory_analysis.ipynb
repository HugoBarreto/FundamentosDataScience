```python
import os
import pandas as pd
import numpy as np
```

```python
datapath = "../data/"
outputs = "../outputs/"
```

```python
df_miraflores = pd.read_csv(os.path.join(datapath,'miraflores-alerts.csv'))
df_montevideo = pd.read_csv(os.path.join(datapath,'montevideo-alerts.csv'))
```

```python
dfs_dict = {}
dfs_dict = {"montevideo": df_montevideo, "miraflores": df_miraflores}
```

```python
print("INFO:\n")
for name,df in dfs_dict.items():
    print(name.upper(), '\n')
    print(df.info(), '\n')
```

```python
print("DESCRIPTION:\n")
for name,df in dfs_dict.items():
    print(name.upper(), '\n')
    print(df.describe(), '\n')
```

```python
print("VARIABLE CATEGORIES:\n")
for name,df in dfs_dict.items():
    print(name.upper(), '\n')
    for cat in df.columns:
        print("Number of levels in category '{0}': \b {1:2.2f} ".format(cat, df_miraflores[cat].unique().size))
    print("\n")
```

```python
print("MISSING DATA:\n")
for name,df in dfs_dict.items():
    print(name.upper(), '\n')
    total = df.isnull().sum().sort_values(ascending=False)
    percent = (df.isnull().sum()/df.isnull().count()).sort_values(ascending=False)
    missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Percent'])
    print(missing_data.head(10))
```
